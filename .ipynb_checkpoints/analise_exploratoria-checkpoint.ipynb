Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

<h2> Estrutura dos dados <h2>

<h2> Importando os dados e tratando para a análise </h2>

```python
# Importando o dataset products e items

products = pd.read_csv(r'csvs\archive\olist_products_dataset.csv')
items = pd.read_csv(r'csvs\archive\olist_order_items_dataset.csv')

# Mesclando os datasets
items = pd.merge(items, products, on='product_id')
```

```python
# Importando o dataset reviews e orders

orders = pd.read_csv(r'csvs\archive\olist_orders_dataset.csv')
reviews = pd.read_csv(r'csvs\archive\olist_order_reviews_dataset.csv')

# Colocando colunas no formato datetime

reviews['review_creation_date'] = pd.to_datetime(reviews['review_creation_date'], format='%Y/%m/%d %H:%M:%S')
reviews['review_answer_timestamp'] = pd.to_datetime(reviews['review_answer_timestamp'], format='%Y/%m/%d %H:%M:%S')
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'], format='%Y/%m/%d %H:%M:%S')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], format='%Y/%m/%d %H:%M:%S')

# Mesclando as tabelas reviews e items
reviews = pd.merge(reviews, items, on='order_id')

# Mesclando as tabelas reviews e orders
reviews = pd.merge(reviews, orders, on='order_id')


# Tirando as colunas que não farão parte da análise
to_drop = [
    'review_id', 
    'order_id', 
    'customer_id', 
    'review_comment_title', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_estimated_delivery_date',
    'order_item_id',
    'product_id',
    'seller_id',
    'shipping_limit_date',
    'price',
    'freight_value',
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]
reviews.drop(columns=to_drop, inplace=True)

```

<h2> Verificando o comportamento padrão dos reviews (de 1 a 5) </h2>

```python
# Criando a porcentagem de reviews por nota
r5 = len(reviews[reviews['review_score'] == 5]) * 100 / len(reviews)
r4 = len(reviews[reviews['review_score'] == 4]) * 100 / len(reviews)
r3 = len(reviews[reviews['review_score'] == 3]) * 100 / len(reviews)
r2 = len(reviews[reviews['review_score'] == 2]) * 100 / len(reviews)
r1 = len(reviews[reviews['review_score'] == 1]) * 100 / len(reviews)

# Analisando quando foi o primeiro e último review
first_dt = reviews['review_creation_date'].min()
last_dt = reviews['review_creation_date'].max()

# Calculando a média geral dos reviews
avg_r = reviews['review_score'].mean()

# Mostrando na tela os resultados de maneira arrendodada
print(len(reviews), 'reviews')
print('First:', first_dt)
print('Last:', last_dt)
print(f'5★: {r5:.1f}%')
print(f'4★: {r4:.1f}%')
print(f'3★: {r3:.1f}%')
print(f'2★: {r2:.1f}%')
print(f'1★: {r1:.1f}%')
print(f'Average: {avg_r:.1f}★')
```

Criando paleta de cores que atenda pessoas daltônicas

```python
sns.set()

# Definindo cores
COLOR_r5 = '#0112EB'
COLOR_r4 = '#F603FF'
COLOR_r3 = '#FAF200'
COLOR_r2 = '#01E6B7'
COLOR_r1 = '#FA4400'
REVIEWS_PALETTE = sns.color_palette((COLOR_r1, COLOR_r2, COLOR_r3, COLOR_r4, COLOR_r5))

# Fundo claro para facilitar a visualização
sns.set_style('darkgrid', {'axes.facecolor': '#eeeeee'})

# Tamanho padrão
resize_plot = lambda: plt.gcf().set_size_inches(12, 5)
```

<h2> Resultado gráfico da distribuição de reviews </h2>

<b>Insights:</b>
1. Notas 1 aparecem mais que notas 2 e 3 somadas
2. A maior parte de reviews é de 5 estrelas

```python
 sns.catplot(
    x='review_score', 
    kind='count', 
    data=reviews, 
    palette=REVIEWS_PALETTE
).set(
    xlabel='Nota dos Reviews', 
    ylabel='Número de reviews',
);
```

Gráfigo retirado de outro trabalho do kaggle com insights importantes:

1. há reviews que foram dados antes da entrega, possivelmente são relativo a problemas com entregas

[Fonte: Customer Reviews Analysis (Brazilian E-Commerce), nilo k](https://www.kaggle.com/krnilo/customer-reviews-analysis-brazilian-e-commerce)



<h2> Relação entre data de entrega e review </h2>

```python
reviews['review_creation_delay'] = (reviews['review_creation_date'] - reviews['order_purchase_timestamp']).dt.days
sns.scatterplot(
    x='order_purchase_timestamp', 
    y='review_creation_delay', 
    hue='review_score', 
    palette= REVIEWS_PALETTE, 
    data=reviews
).set(
    xlabel='Purchase Date', 
    ylabel='Review Creation Delay (days)',
    xlim=(datetime(2016, 8, 1), datetime(2018, 12, 31))
);
resize_plot()
```
